```python
import numpy as np
import pandas
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Making the data:
np.random.seed(42)
size = int(input('Define the number of trials: '))
flips = np.random.choice(2, size=size)
means = np.empty_like(flips, dtype=np.float64)
for i in range(flips.size):
    means[i] = np.mean(flips[:i+1])

print(f'Mean value for {size} trials is {means[size-1]}')

# Making the plot:
plt.plot(means)
plt.xlabel('Number of trials')
plt.ylabel('Mean value')
plt.show()
```
