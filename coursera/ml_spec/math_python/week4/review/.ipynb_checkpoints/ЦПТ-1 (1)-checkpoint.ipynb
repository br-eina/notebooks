```python
import matplotlib
import matplotlib.pyplot as plt
from scipy.stats import expon
import numpy as np
import scipy.stats as sts
from matplotlib.pyplot import plot, hist, xlabel, legend
from numpy import linspace, array, mean
```

```python
# Я выбрала распределение Пуассона.
x = linspace(-10, 30, 41)
rv = sts.poisson(10)
pois = rv.rvs(1000)
pmf = rv.pmf(x)

hist(pois, bins=15, density=True) #строим гистограмму

plot(x, pmf, 'r', label='Пуассонновское распределение') #строим теоретическую плотность поверх гистограммы
xlabel('x')
legend(fontsize='small')
```

```python
#Выборка объема 10.
data = list()
for i in range(0, 1000):
    data.append(rv.rvs(10))
data = array(data)

mean_data = list()
for sample in data:
    mean_data.append(mean(sample))
hist(mean_data, bins=20, density=True)

norm_rv = sts.norm(10, (10/15)**0.5)
x_10 = linspace(5, 15, 100)
norm_pdf = norm_rv.pdf(x_10)
plot(x_10, norm_pdf, label='Нормальное распределение')

xlabel('x')
legend(fontsize='small')
```
