```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline
```

# 1. Гистограмма и плотность распределения экспоненциального распределения

```python
rv = sts.expon()

# Генерация выборки и построение гистограммы
val = rv.rvs(size=1000)
plt.hist(val, density=True)

# Плотность распределения
x = np.linspace(0,10,1000)
pdf = rv.pdf(x)
plt.plot(x, pdf, 'r-', lw=3,  label='плотность распределения')

plt.legend(loc='best')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
```

# 2. Гистограмма выборочных средних размера 5 и плотность нормального распределения

```python
size = 5
mu, sigma = 1, 1 # mean and standard deviation

val = []
for i in range(1000):
    val.append(np.mean(sts.expon.rvs(size = size)))
    
plt.hist(val, 50, density=True)

# Плотность нормального распределения
x = np.linspace(-5,5,1000)
plt.plot(x, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (x - mu)**2 / (2 * sigma**2) ), linewidth=2, color='r')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show() 
```
