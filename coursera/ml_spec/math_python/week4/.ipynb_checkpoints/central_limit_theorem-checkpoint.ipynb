```python
# Импортирование библиотек
%matplotlib inline
import math
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import pandas as pd
plt.style.use('ggplot') # Изменение стиля графиков
```

```python
# Генерирование выборки случайной величины
rv_distrib = sts.semicircular()
sample = rv_distrib.rvs(1000)
# Генерирование значений плотности распределения
x = np.linspace(-2,2,100)
pdf = rv_distrib.pdf(x)
# Построение гистограммы и плотности распределения
plt.figure(figsize=(10,6))
plt.plot(x, pdf, label='theoretical pdf', lw=3)
plt.hist(sample, bins=15, density=True, label='histogram')
plt.legend(fontsize=15)
plt.ylabel('$f(x)$', fontsize=15)
plt.xlabel('$x$', fontsize=15)
plt.title('Semicircle distribution')
plt.show()
```

```python
plt.figure(figsize=(20, 6))
# Матожидание и дисперсия исходной случайной величины
# https://en.wikipedia.org/wiki/Wigner_semicircle_distribution
base_distr_mean = 0
base_distr_var = 1/4
# Размеры и количество выборок
sample_sizes = [5, 10, 50]
samples_number = 1000
# Генерация выборок средних и их визуализация
for ind, sample_size in enumerate(sample_sizes):
    means = np.empty(samples_number)
    for i in range(samples_number):
        sample = rv_distrib.rvs(sample_size)
        means[i] = np.mean(sample)
    # Плотность нормального распределения с необходимыми параметрами под текущее распределение средних
    norm_distr = sts.norm(base_distr_mean, math.sqrt(base_distr_var/sample_size))
    x = np.linspace(-1, 1, 200)
    pdf = norm_distr.pdf(x)
    # Визуализация распределения средних
    plt.subplot(1, len(sample_sizes), ind+1)
    plt.plot(x, pdf, label='theoretical pdf', lw=3)
    plt.hist(means, bins=20, density=True, label='histogram')
    plt.legend(fontsize=12)
    plt.ylabel('$f(x)$', fontsize=15)
    plt.xlabel('$x$', fontsize=15)
    plt.title(f'Sample size $n={sample_size}$\n'
              f'$mean_n={"%.2E" % np.mean(means)}$')

plt.show()
```

В связи с тем, что исходное распределение случайной величины не скошено и обладает малой дисперсией, распределение средних по выборкам хорошо описывается нормальным распределением даже при малых размерах выборки *n*. Однако четко видно, что исходное распределение отличается от нормального.

Тем не менее, при увеличении размера выборки точность аппроксимации нормальным распределением растет. Дисперсия нормального распределения уменьшается, и средние сильнее концентрируются вокруг выборочного среднего. Выборочное среднее также приближается к матожиданию исходной случайной величины.
