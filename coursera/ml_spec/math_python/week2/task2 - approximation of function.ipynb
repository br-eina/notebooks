```python
%matplotlib inline
import numpy as np
import math
from matplotlib import pylab as plt
```

```python
def f(x):
    return np.sin(x/5)*np.exp(x/10) + 5*np.exp(-x/2)

x = np.arange(1, 15, 0.1)
plt.plot(x, f(x))
plt.show()
```

Зададим многочлен 1-ой степени:

```python
def polynom(x, w0, w1):
    return w0 + w1*x

w0 + w1x1 = f(x1)
w0 + w1x2 = f(x2)


# Find coefficients w_0 and w_1
x1, x2 = 1, 15
A = np.array([[1, x1], [1, x2]], dtype=float)
b = np.array([f(x1), f(x2)], dtype=float)
w0, w1 = np.linalg.solve(A, b)

plt.plot(x, polynom(x, w0, w1))
plt.show()
```

Зададим многочлен 2-ой степени:

```python
def polynom(x, w0, w1, w2):
    return w0 + w1*x + w2*(x**2)

# Find coefficients w_0, w_1 and w_2
x1, x2, x3 = 1, 8, 15
A = np.array([[1, x1, x1**2], 
              [1, x2, x2**2],
              [1, x3, x3**2]], dtype=float)
b = np.array([f(x1), f(x2), f(x3)], dtype=float)
w0, w1, w2 = np.linalg.solve(A, b)

plt.plot(x, polynom(x, w0, w1, w2))
plt.show()
```

Зададим многочлен 3-ей степени:

```python
def polynom(x, w0, w1, w2, w3):
    return w0 + w1*x + w2*(x**2) + w3*(x**3)

# Find coefficients w_0, w_1 and w_2
x1, x2, x3, x4 = 1, 4, 10, 15
A = np.array([[1, x1, x1**2, x1**3], 
              [1, x2, x2**2, x2**3],
              [1, x3, x3**2, x3**3],
              [1, x4, x4**2, x4**3]], dtype=float)
b = np.array([f(x1), f(x2), f(x3), f(x4)], dtype=float)
w0, w1, w2, w3 = np.linalg.solve(A, b)

plt.plot(x, polynom(x, w0, w1, w2, w3))
plt.plot(x, f(x))
plt.show()
w0, w1, w2, w3
```
