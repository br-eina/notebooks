**Binary Search**: Search a sorted array by repeatedly dividing the search interval in half. Begin with an interval covering the whole array. If the value of the search key is less than the item in the middle of the interval, narrow the interval to the lower half. Otherwise narrow it to the upper half. Repeatedly check until the value is found or the interval is empty.

Binary search is a fast search algorithm with run-time complexity of Ο(log n).

```python
def binary_search(arr, l, r, x):
    while l <= r:
        mid = l + (r - l) // 2
        
        # Check if x is present at mid:
        if arr[mid] == x:
            return mid
        
        # If x if greater, search only right half:
        if arr[mid] < x:
            l = mid + 1
        
        # If x is smaller, search only left half:
        if arr[mid] > x:
            r = mid - 1
        
    # If we are here, the element is not present
    return -1    
```

```python
array = [2, 3, 4, 5, 7, 9, 11]
x = 11
```

```python
result = binary_search(array, 0, len(array) - 1, x)

if result != -1:
    print("Element is present at index {}".format(result))
else:
    print("There is no such element in array")
```
